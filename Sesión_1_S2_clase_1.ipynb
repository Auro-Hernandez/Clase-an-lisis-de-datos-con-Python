Las listas en Python tienen las siguientes características:

Son ordenadas: cada elemento dentro de la lista está indexado y mantiene el orden definido (al primer elemento se le asigna el valor de 0).

Permiten elementos duplicados: las listas pueden tener elementos con el mismo valor.

Las listas pueden anidar, es decir, contener una lista dentro de la otra.

Son mutables: podemos editar, eliminar o agregar elementos en una lista existente (nombre de la lista.append(), nombre de la lista.remove(),lista1[x] = y)

Para cambiar el valor de una lista:

```python
lista1[1]=15 #Se cambia el valor de la posición [1] al valor 15
```
