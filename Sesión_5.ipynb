Sistema de ecuaciones

1. Identifica la matriz de coeficientes (A)

```python
import numpy as np
```

```python
A=np.array([[2,4,6],[4,5,6],[3,1,-2]])
```

2. Calcula el determinante principal (D o det(A))

```python
D=np.linalg.det(A)
D
```

3. Crea matrices modificadas para cada variable


*   Para encontrar X:reemplaza la primeta columna de A con el vector de términos independientes.
*   Para encontrar Y: reemplaza la segunda columna
Para encontrar Z: reemplaza la tercera columan (y así sucesivamente)



```python
x=np.array([[18,4,6],[24,5,6],[4,1,-2]])
```

4. Calcula el determinante de cada matriz modificada
Dx, Dy, Dz, etc.

```python
Dx=np.linalg.det(x)
Dx
```

5. Aplica la fórmula Cramer
x=Dx/D
y=Dy/D
z=Dz/D

```python
x=Dx/D
x
```

```python
b=np.array([18,24,4]) #Comprobación

x=np.linalg.solve(A,b)
x
```

```python
np.dot(A,x)
```
