**La codificación**

Es la forma en que las computadoras convierten texto en números para poder guardarlo y leerlo. Cada carácter (letam número, símbolo) tiene un código numérico.

**Archivos TXT**

Un archivo txt se puede leer con read_csv() porque ambos son solo texto delimitado. Panda no se fija en la extensión, sino en como están separados los datos. Si el txt tiene comas, punto y coma, tabulaciones o espacios, read_csv() sabe interpretarlo como una tabla.

read_csv() lee CSVs:


*   .txt
*   .data
*   .tsv
*   archivos sin extensión
*   cualquier archivo de texto que tenga delimitadores

# **Convertir datos de un tipo a otro**

```python
# Cambiar tipo de dato
df["fecha"] = pd.to_datetime(df["fecha"]) # Convertir a fecha
df["edad"] = df["edad"].astype(int)
df['edad'] = df['edad'].astype(float)
df['nombre'] = df['nombre'].astype('category')

# Seleccionar solo columnas numéricas
df.select_dtypes(include=['int64', 'float64'])

# Seleccionar solo columnas de texto
df.select_dtypes(include=['object'])

# Contar tipos de datos
df.dtypes.value_counts()
```

# **Renombrar columnas**

```python
df.rename(columns={"Temp": "Temperatura"}, inplace=True) #Renombra la columna "Temp" a "Temperatura" y modifica el DataFrame
```

```python
# Renombrar varias columnas a la vez
df.rename(columns={
    "Temp": "Temperatura",
    "Hum": "Humedad",
    "Pres": "Presión"
}, inplace=True)

print(df.columns.tolist())
# ['Temperatura', 'Humedad', 'Presión']
```

```python
# Renombrar TODAS las columnas (sobrescribir lista completa)
df.columns = ['Temperatura', 'Humedad', 'Presión']

# Usando la función upper() (convertir a mayúsculas) a cada nombre de columna.
df.rename(columns=str.upper, inplace=True)
# Temp → TEMP, Hum → HUM

df.columns = df.columns.str.lower()  # Minúsculas

# Añadir prefijo o sufijo
df.add_prefix('col_')   # col_Temp, col_Hum
df.add_suffix('_data')  # Temp_data, Hum_data

# Reemplazar espacios o caracteres
df.columns = df.columns.str.replace(' ', '_') #De espacios a guiones
# 'Nombre Completo' → 'Nombre_Completo'
```
