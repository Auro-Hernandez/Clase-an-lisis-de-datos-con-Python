Un módulo es un archivo que contiene código Python (funciones, variables o clases) que puedes importar en otros programas.

# **if, elif, else**

Permiten que el programa time decisiones según condiciones.

if: comprueba una condición, se ejecuta si la condición es verdadera.

elif: se ejecuta si la anterior fue falsa pero esta esta es verdaderañ

else: se ejecuta si ninguna condición anterior fue verdadera, no lleva condición.

```python
x=56
if x<0:

  print('x es negativo')
```

```python
x=18
if x<0:
  print('x es negativo')#sangría 4 espacios
elif x==0:
    print('x es igual a cero')
else:
      print('x es positivo')
```

```python
n=int(input('Ingresa un número:'))#int hace que la sentencia se reconozca como un número
if n>0:
  print('El número es positivo')
elif n<0:
  print('El número es negativo')
else:
  print('El número es cero')
```
