**MÉTODO ACF: FUNCIÓN DE AUTOCORRELACIÓN, NOS AYUDA A DECIR SI NUESTOS DATOS TIENEN ALGUNA CORRELACIÓN**

Nos dicen que tan relacionados están los datos entre ellos

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df_15['PM10'], lags=150)
plt.show()
```

Si hay una descripción con ondulaciones definidas tenemos estacionalidad, por ejemplo se repite un ciclo cada semana

Si la correlación disminuye gradualmente pero se mantiene alta, podemos considerar posible no estacionariedad

La banda azul representa los intervalos de confianza

```python
df_15.columns #Nombre de las columnas
```

```python
#Lista de columans/variables a analizar
variables=['CO', 'NO', 'NO2', 'NOX', 'O3', 'PM10', 'PM25', 'SO2']

#Crea una figura con subplots
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(12, 10))
axes=axs.flatten()

#Grafica ACF para cada variable

for i, var in enumerate(variables):
  plot_acf(df_15[var].dropna(), lags=150, ax=axes[i])
  axes[i].set_title(f'ACF para {var}')

plt.tight_layout()
plt.show()
```

```python
#Si mi serie se va a cero es estacionaria
```

CO: tiene ondulaciones, por tanto tiene posible estacionalidad y la serie es estacionaria.
NO:tiene ondulaciones débiles, por tanto tiene posible estacionalidad y la serie es estacionaria.
NO2:  tiene ondulaciones débiles, por tanto tiene posible estacionalidad, pero sí se va a cero, por tanto es estacionaria.
Nox: tiene ondulaciones, por tanto tiene posible estacionalidad y la serie estacionaria.
O3: No tiene ondulaciones, por tanto no tiene posible estacionalidad, pero sí se va a cero, por tanto es estacionaria.

# **Método de descomposición de tendencias STL**

```python
from statsmodels.tsa.seasonal import STL

stl=STL(df_15['NOX'],period=7) #periodo 7 estacionalidad semanal
res=stl.fit()

#Graficar los componentes

fig=res.plot()

#Rotar etiquetas del eje x en todos los subplots

for ax in fig.axes:
    ax.tick_params(axis='x', labelrotation=45) #Rotar 45 grados para una mejor legibilidad

plt.tight_layout() #Ajusta el espacio entre subplots automáticamente
plt.show()
```

Season: 25 es el valor máximo, y el mínimo -25, por tanto el rango sería 0.

Residuo= 25 es el valor máximo, y el mínimo -25, por tanto el rango sería 0.
