# **Uso de condicionantes**

Escriba un código que determine el regímen de flujo para cierto valor de Re.

```python
import math
```

```python
rho=1000 #densidad [Kg/m3]
v=0.3    #velocidad [m/s]
D=0.05   #Diámetro [m]
mu=0.001 #Viscosidad [Pa*s]
```

```python
Re=(rho*v*D)/(mu) # 1. Escribimos la ecuación
print(f'Re={Re:.2f}') # 2. Invocamos la ecuación con prin (f'función a utilizar:.num. de float')
if Re<2100:  #Usar los condicionantes, recordar siempe las sangrías
 print('Flujo láminar')
elif 2100<=Re<=4000:
 print('Flujo es transitorio')
elif Re>4000:
  print('Flujo turbulento')

```

# **Bucles**

Un bucle (o ciclo) es una estructura que permite repetir instrucciones varias veces sin tener que escribir el mismo códgio una y otra vez.

Un bucle le dice al programa que repita una acción mientras cumpla la condición o durante cierto número de veces.

La **función range** genera una secuencia de enteros espaciados por igual.

```python
range(20)
```

```python
list(range(20)) #para volver lista
```

```python
for i in range(5): #for para ciclo, "in" respresenta "hasta"
  print('número:',i) # Estoy llamando los primeros 5 "números"
```

```python
L=[1,2,None,4,None,5] # 1. Crea una variable que contenga una lista
```

```python
list(L) #Invoca la lista
```

```python
for i in L: # Le da el formato de lista
  print(i)
```

```python
suma=0 #Suma los primeros 1000 números
for i in range(1,1001):
  suma+=i
suma
```

# **Bucle while**

Se usa cuando no sabemos cuántas veces se repetirá, solo sebemos que queremos repetir mientras se cumpla una condición.

```python
x=0
while x<5:
  print('x vale',x) #Decirle que queremos que imprima ('x vale'), y la variable que debe imprimir
  x+=1 #Para que vaya imprimiento el valor siguiente, todo lo que vaya dentro de un ciclo tiene que ir con identación
```

# **Bucles for**

Se usan cuando sabemos cuántas veces queremos repetir algo. Repite una acción por cada elemento de una secuencia (como una lista, un rangoo una cadena de texto).

```python
for i in range(1,10):#Para el rango de 1-10.....
  if i==5: #....Detiene el ciclo al llegar a 5
   break
  print(i) #print para mostrar la acción
```

```python
L=[1,2,None,4,None,5]
```

```python
suma=0
for i in L:
  if i is None:continue #continuo salta una línea
  suma +=i #+= para que me de la suma de todos lo valores
suma
```
