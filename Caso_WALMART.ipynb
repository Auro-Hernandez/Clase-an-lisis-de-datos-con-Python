**1. Siempre importar todas las paqueterias que se van a utilizar**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**2.Insertar los datos, siempre asignándoles un nombre**

```python
ruta='https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/WALMEX.csv'
```

3. Leer/invocar los datos

```python
df=pd.read_csv(ruta)
df.head()
```

```python
df.info()
```

4. Convertir los datos una serie de tiempo

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df
```

**5. Indexar los datos**

```python
df=df.set_index('Date')
df
```

```python
df.idxmax() #.idxmax devuelve el valor del indice en el que alcaqnza el valor máximo
```

```python
df.idxmin() #.idxmin devuelve el valor del indice en el que alcaqnza el valor mínimo
```

**6. Graficar**

```python
plt.figure(figsize=(16,8),dpi=150) #Las longitudes están en pulg

#Línea principal del precio de cierre
plt.plot(df.index,df['Close'], lw=2.5, color='black', label='Precio de cierre') #lw va a variar el grosor de la gráfica, primero se añadé el eje horizontal y vertical, y depués las especificaciones del gráfico

#Etiquetas de títulos

plt.xlabel('Fecha', fontsize=20, labelpad=10) #fontsize para tamaño de la letra, labelpad tamaño de la separación
plt.ylabel('Precio de cierre', fontsize=20, labelpad=10)
plt.title('Precio de cierre WALMEX', fontsize=22, fontweight='bold', pad=15)

#Configuración del gráfico
plt.grid(alpha=0.3)
plt.tick_params(labelsize=20)
plt.legend(fontsize=25)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Preparación de los datos y encontrar Extremos (Max y Min)
# Asumiendo que df ya está cargado y tiene las columnas 'Close' y un índice de fecha.
# --- ESTO ES LO QUE HACE QUE SEA MÁS CIENTÍFICO Y ANALÍTICO ---
idx_max = df['Close'].idxmax()
val_max = df['Close'].max()
idx_min = df['Close'].idxmin()
val_min = df['Close'].min()

# 2. Configuración de la Figura
plt.figure(figsize=(18, 9), dpi=200) # Mayor resolución (DPI) para calidad científica

# 3. Gráfico de la Serie Temporal
plt.plot(
    df.index, df['Close'],
    lw=2.5,
    color='#00529B', # Color azul más formal
    label='Precio de Cierre ($)',
    alpha=0.8 # Transparencia sutil
)

# 4. Marcado y Anotación de los Valores Extremos
# --- PUNTO CRÍTICO PARA LA MEJORA EXPLICATIVA ---

# Valor Máximo
plt.scatter(idx_max, val_max, color='red', s=150, zorder=5, label=f'Máximo: {val_max:.2f}')
plt.annotate(
    f'Máximo: ${val_max:.2f}', # Texto de la anotación
    (idx_max, val_max),       # Posición del punto
    textcoords="offset points",
    xytext=(-30, 10),         # Desplazamiento del texto
    ha='center',              # Alineación
    fontsize=14,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='red')
)

# Valor Mínimo
plt.scatter(idx_min, val_min, color='green', s=150, zorder=5, label=f'Mínimo: {val_min:.2f}')
plt.annotate(
    f'Mínimo: ${val_min:.2f}',
    (idx_min, val_min),
    textcoords="offset points",
    xytext=(-30, -25),
    ha='center',
    fontsize=14,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='green')
)

# 5. Etiquetas y Títulos Mejorados
plt.xlabel('Fecha de Observación', fontsize=18, labelpad=15)
plt.ylabel('Precio de Cierre (MXN)', fontsize=18, labelpad=15)
plt.title(
    'Análisis de la Serie Temporal del Precio de Cierre de WALMEX',
    fontsize=22,
    fontweight='bold',
    pad=20
)

# 6. Configuración Estilística (Alineada a estándares científicos)
plt.grid(
    True,
    linestyle='--',
    alpha=0.5,
    which='both' # Mostrar cuadrícula para ejes mayores y menores
)
plt.tick_params(labelsize=14)
plt.legend(loc='upper left', fontsize=16, frameon=True, shadow=True) # Mejor ubicación y estilo de leyenda
plt.tight_layout() # Ajusta automáticamente los parámetros de la subtrama para que quepa toda la figura
plt.show()
```
