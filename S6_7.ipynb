```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Preparación de Datos ---

# Generar datos de ejemplo
np.random.seed(42) # Para reproducibilidad
x = np.linspace(0, 10, 50)
y = 3 * np.sin(x) + np.random.normal(0, 0.5, 50) # Patrón con ruido

# Encontrar los valores mínimo y máximo
y_min = np.min(y)
x_min = x[np.argmin(y)]
y_max = np.max(y)
x_max = x[np.argmax(y)]

# --- 2. Configuración del Estilo ---

# Aplicar un estilo predefinido para un look limpio
plt.style.use('seaborn-v0_8-whitegrid') # 'seaborn-v0_8-whitegrid' es limpio y tiene líneas de cuadrícula útiles
# Alternativas: 'ggplot', 'fivethirtyeight', 'default'

# --- 3. Creación de la Figura y los Ejes ---

# Crear la figura y los ejes usando la interfaz orientada a objetos (recomendada)
fig, ax = plt.subplots(figsize=(10, 6)) # figsize para un buen tamaño

# --- 4. Trazado Principal ---

# Trazar la línea principal
ax.plot(x, y,
        color='steelblue',      # Color de la línea
        linewidth=2,            # Grosor de la línea
        label='Valores de la Serie')

# --- 5. Resaltar Mínimo y Máximo ---

# Punto Mínimo
ax.scatter(x_min, y_min,
           color='red',         # Color del marcador
           s=100,               # Tamaño del marcador (más grande)
           zorder=5,            # Asegura que el punto esté encima de la línea
           label=f'Mínimo ({y_min:.2f})')
ax.annotate(f'Min: {y_min:.2f}', # Añadir etiqueta de texto
            (x_min, y_min),
            textcoords="offset points",
            xytext=(10, -15),   # Desplazamiento de la etiqueta
            ha='left',
            color='red',
            fontweight='bold')

# Punto Máximo
ax.scatter(x_max, y_max,
           color='green',       # Color del marcador
           s=100,
           zorder=5,
           label=f'Máximo ({y_max:.2f})')
ax.annotate(f'Max: {y_max:.2f}',
            (x_max, y_max),
            textcoords="offset points",
            xytext=(-30, 10),
            ha='right',
            color='green',
            fontweight='bold')

# Líneas de referencia horizontal opcionales
ax.axhline(y_min, color='red', linestyle='--', linewidth=0.8, alpha=0.5)
ax.axhline(y_max, color='green', linestyle='--', linewidth=0.8, alpha=0.5)


# --- 6. Configuración de Ejes y Títulos (Limpieza y Estética) ---

# Título y etiquetas
ax.set_title('Análisis de Tendencia con Puntos Extremos', fontsize=16, fontweight='bold')
ax.set_xlabel('Tiempo (Eje X)', fontsize=12)
ax.set_ylabel('Magnitud del Valor (Eje Y)', fontsize=12)

# Mejorar los ticks de los ejes (opcional)
ax.tick_params(axis='both', which='major', labelsize=10)

# Leyenda
ax.legend(loc='best', frameon=True) # frameon=True para un borde alrededor de la leyenda

# Eliminar el borde superior y derecho (Spines) para un look más limpio
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(0.5)

# Ajustar diseño (por si los títulos se superponen)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

Uso del Estilo (plt.style.use()): Aplicar un estilo como 'seaborn-v0_8-whitegrid' cambia automáticamente muchos parámetros por defecto (fuentes, colores, fondo), resultando en un aspecto profesional sin esfuerzo.

Interfaz Orientada a Objetos (fig, ax = plt.subplots()): Usar ax.plot(), ax.set_title(), etc., en lugar de plt.plot(), plt.title() da mayor control sobre los elementos del gráfico, especialmente útil en gráficos complejos o con múltiples subplots.

Resaltar Min/Max (ax.scatter y ax.annotate):

ax.scatter() se usa para dibujar los puntos con un tamaño (s=100) y color específicos.

ax.annotate() añade una etiqueta de texto cerca del punto, especificando el valor exacto para facilitar la lectura.

zorder=5 asegura que los puntos y las anotaciones se dibujen encima de la línea principal.

ax.axhline() añade líneas de referencia horizontales sutiles para el máximo y el mínimo.

Limpieza de Ejes (ax.spines): Ocultar los bordes superior y derecho (ax.spines['right'].set_visible(False)) es una técnica común para hacer el gráfico menos recargado y dirigir la atención a los datos.

# Coeficiente de curtosis

La curtosis describe el grado en que las colas de una distribución difieren de las colas de una distribución normal. Mide qué tan afilada o achatada es la distribución en comparación con la normal

```python
# Calcular la curtosis de Fisher (exceso de curtosis) para cada contaminante
curtosis = contam.kurt()

# Mostrar los resultados
print("\nCurtosis de Fisher:\n")
print(curtosis)
```

```python

# --- 1. Calcular curtosis de Fisher ---
curtosis = contam.kurt()
# --- 2. Crear los histogramas ---
columnas = contam.columns

plt.figure(figsize=(15, 8))  # tamaño del lienzo

for i, col in enumerate(columnas, 1):
    plt.subplot(2, 3, i)  # 2 filas, 3 columnas de gráficos
    plt.hist(contam[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f"{col}\nCurtosis: {curtosis[col]:.2f}", fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Frecuencia')


plt.suptitle("Distribución de contaminantes y su curtosis (Fisher)", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Configuración global del estilo para formalidad
plt.style.use('seaborn-v0_8-whitegrid')

# --- Argumentos clave para la estética científica ---
# These parameters should be passed to scatter_kws within plot_kws when kind='reg'
scatter_style_params = {
    's': 50,              # Tamaño de los marcadores
    'alpha': 0.7,         # Transparencia para ver el solapamiento
    'edgecolor': 'k',     # Borde negro para claridad
    'linewidths': 0.5      # Grosor del borde (corrected from linewidth to linewidths)
}

# Arguments for the diagonal (KDE)
diag_params = {
    'fill': True,
    'alpha': 0.5
}

# --- Generación del pairplot mejorado ---
g = sns.pairplot(
    contam,
    # 1. Elementos Estadísticos y Estilo
    kind='reg',           # Usa regresión lineal para mostrar tendencias
    diag_kind='kde',      # Usa estimación de densidad de kernel en la diagonal
    corner=True,          # Mantener solo el triángulo inferior

    # 2. Desagregación (si aplica)
    # hue='Grupo',          # Columna categórica para diferenciar grupos (ajusta este nombre)
    palette='viridis',    # Mapa de color perceptualmente uniforme

    # 3. Personalización Estética (Diccionarios de kws)
    plot_kws={'scatter_kws': scatter_style_params}, # Pass scatter_style_params here
    diag_kws=diag_params
)

# --- Personalización de Ejes y Título (Formalidad) ---
# El pairplot es un objeto FacetGrid, que requiere iteración
# Título general
g.fig.suptitle("Matriz de Dispersión y Regresión: Análisis Multivariado de Contaminantes",
               fontsize=16, y=1.02)

# Opcional: Ajuste fino de etiquetas de ejes (si es necesario cambiar nombres largos)
for i, ax in enumerate(g.axes.flat):
    if ax: # Solo en los subgráficos existentes
        # Rotar etiquetas X para evitar solapamiento
        ax.set_xlabel(ax.get_xlabel(), rotation=45, ha='right')
        # Añadir cuadrícula específica para los subgráficos
        ax.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout(rect=[0, 0, 1, 0.98]) # Ajuste para el título
plt.show()
```

Person vs Sperman

```python
from scipy.stats import spearmanr, pearsonr

# Datos de ejemplo (no lineales, pero monótonos)
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 6, 8, 10, 12, 14, 16, 30]  # crece en general, pero con un valor extremo al final

# Crear un DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Calcular las correlaciones
r_pearson, p_pearson = pearsonr(df['X'], df['Y'])
r_spearman, p_spearman = spearmanr(df['X'], df['Y'])

print(f"Coeficiente de Pearson:  r = {r_pearson:.3f}, p = {p_pearson:.4f}")
print(f"Coeficiente de Spearman:  ρ = {r_spearman:.3f}, p = {p_spearman:.4f}")
```

```python

plt.scatter(df['X'], df['Y'], color='royalblue')
plt.title('Relación entre X y Y (monótona pero no lineal)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(alpha=0.3)
plt.show()
```
